```python
category  = movies.Genre.cat.categories
```

```python
movie_genre = []
labels = []
for gen in category:
    movie_genre.append(movies[movies.Genre == gen ].BudgetMillion)
    labels.append(gen)
    
plt.hist(movie_genre, stacked = True,label = labels ) 
plt.legend()
plt.show()
```

#### Controlling axes


```python
g = sns.FacetGrid(movies, row = "Genre", col = 'Year',hue = 'Genre')
kws  = dict(s=50,linewidth = 0.5, edgecolor = 'black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
g.set(xlim = (0,100), ylim = (0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c= 'gray',ls= "--")
g.add_legend()    
```
